```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow,imread,imsave
```

```python
img = cv2.imread('house.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
```

```python
#sobel
sobelx = cv2.Sobel(img_gaussian,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img_gaussian,cv2.CV_64F,0,1,ksize=5)
sobel=sobelx + sobely
```

```python
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
img_prewitt=img_prewittx+ img_prewitty
```

```python
#canny
canny = cv2.Canny(img,100,200)
```

```python
#Robert
kernellx = np.array( [[1, 0 ],
                             [0,-1 ]] )

kernelly = np.array( [[ 0, 1 ],
                             [ -1, 0 ]] )

robertx = cv2.filter2D( img_gaussian,-1, kernellx )
roberty = cv2.filter2D( img_gaussian,-1, kernelly )
robert=robertx+roberty
```

```python
plt.figure(figsize=(20,8))
plt.subplot(321),plt.imshow(img),plt.title('Original'),plt.axis('off')
plt.subplot(322),plt.imshow(sobel),plt.title('Sobel'),plt.axis('off')
plt.subplot(323),plt.imshow(img_prewitt),plt.title('Prewitt'),plt.axis('off')
plt.subplot(324),plt.imshow(canny),plt.title('Canny'),plt.axis('off')
plt.subplot(325),plt.imshow(robert),plt.title('Robert'),plt.axis('off')
```
